```python
from bs4 import BeautifulSoup
import requests

all_hyperlinks = []

for page_num in range(1, 3):  #this way it goes thru both pages without having two diff links
    url = f"https://www.lse.ac.uk/student-life/accommodation/search-accommodation?collection=lse-accommodation&pageIndex={page_num}&sort=metaavailability"
    try:
        response = requests.get(url)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')
        accommodation_titles = soup.find_all('h2', class_='card__title')
        hyperlinks = []
        for title in accommodation_titles:
            hyperlink = title.find('a')['href']
            hyperlinks.append(hyperlink)
        all_hyperlinks.extend(hyperlinks)

    except Exception as e:
        print("An error occurred:", e)

print(all_hyperlinks)
```

```python
from bs4 import BeautifulSoup
import requests

# Specify the URL of the page you want to scrape
url = "https://www.lse.ac.uk/student-life/accommodation/halls/rosebery-hall"

try:
    response = requests.get(url)
    response.raise_for_status()  # Raises an exception for 4XX or 5XX responses
    soup = BeautifulSoup(response.text, 'html.parser')
    
    rooms_content = []  # This will store the content of <p> and <h2> tags from each 'rooms' div
    
    # Find all <div> tags with the class 'rooms'
    rooms_divs = soup.find_all('div', class_='rooms')
    for div in rooms_divs:
        # For each 'rooms' div, find all <p> and <h2> tags and extract their text
        content = {"h2": [], "p": []}
        for h2 in div.find_all('h2'):
            content["h2"].append(h2.get_text(strip=True))
        for p in div.find_all('p'):
            content["p"].append(p.get_text(strip=True))
        rooms_content.append(content)

except Exception as e:
    print("An error occurred:", e)

# If you want to print the extracted content:
for room in rooms_content:
    if room["h2"]:  # Check if there is any <h2> content
        print("H2:", ", ".join(room["h2"]))
    if room["p"]:  # Check if there is any <p> content
        print("P:", ", ".join(room["p"]))
    print("------")  # Separator for clarity
```

```python
from bs4 import BeautifulSoup
import requests

# Specify the URL of the page you want to scrape
url = "https://www.lse.ac.uk/student-life/accommodation/halls/rosebery-hall"

try:
    response = requests.get(url)
    response.raise_for_status()  # Raises an exception for 4XX or 5XX responses
    soup = BeautifulSoup(response.text, 'html.parser')
    
    rooms_content = []  # This will store the content of <p> and <h2> tags from each 'rooms' div as strings
    
    # Find all <div> tags with the class 'rooms'
    rooms_divs = soup.find_all('div', class_='rooms')
    for div in rooms_divs:
        # Initialize a dictionary to store the content as strings
        content = {"h2": "", "p": ""}
        # Concatenate all <h2> tag texts into a single string
        content["h2"] = " ".join(h2.get_text(strip=True) for h2 in div.find_all('h2'))
        # Concatenate all <p> tag texts into a single string
        content["p"] = " ".join(p.get_text(strip=True) for p in div.find_all('p'))
        rooms_content.append(content)

except Exception as e:
    print("An error occurred:", e)

# Print the extracted content:
for room in rooms_content:
    if room["h2"]:  # Check if there is any <h2> content
        print("H2:", room["h2"])
    if room["p"]:  # Check if there is any <p> content
        print("P:", room["p"])
    print("------")  # Separator for clarity
```

```python
from bs4 import BeautifulSoup
import requests

# Specify the URL of the page you want to scrape
url = "https://www.lse.ac.uk/student-life/accommodation/halls/rosebery-hall"

# String to hold all the room content
all_rooms_content = ""

try:
    response = requests.get(url)
    response.raise_for_status()  # Raises an exception for 4XX or 5XX responses
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find all <div> tags with the class 'rooms'
    rooms_divs = soup.find_all('div', class_='rooms')
    for div in rooms_divs:
        # Initialize a string to store the concatenated content
        room_content = ""
        # Concatenate all <h2> tag texts into a single string
        room_content += "H2: " + " ".join(h2.get_text(strip=True) for h2 in div.find_all('h2'))
        # Concatenate all <p> tag texts into a single string
        room_content += " P: " + " ".join(p.get_text(strip=True) for p in div.find_all('p'))
        # Add this room's content to the overall content string with a separator
        all_rooms_content += room_content + " ------ "
    
except Exception as e:
    print("An error occurred:", e)

# Print the entire content as one string
print(all_rooms_content)
```

```python
import pandas as pd
import re

# Assuming rooms_content is a list of descriptions as you provided
rooms_content = [
    "H2: Single room, Twin room, Twin en suite room P: Shared bathroom, 39-week contract, £247.80-£263.90p/wk, Catered, No Availability, Contract cost:Rosebery Wing: £9,996.00Myddelton Wing: £9,628.80 to £10,254.40, Contract length: 39 weeks Size approx: 10.5m² to 11m², Room with single bed, wash basin, wardrobe, desk and chair, lamp and storage. Duvet and pillows provided.Also see:What we provide in each hall[PDF]Table of bed sizes[PDF]Contract lengths and dates",
    "Shared bathroom, 39-week contract, £168.70p/wk, Catered, No Availability, Contract cost: £6,555.20 Contract length: 39 weeks, Size approx: 15.5m² to 17.5m², Room with two single beds and a wash basin. A wardrobe, bedside table, desk and chair, lamp and storage are provided per person. Duvet and pillows provided.Also see:What we provide in each hall[PDF]Table of bed sizes[PDF]Contract lengths and dates",
    "Private bathroom, 39-week contract, £185.50p/wk, Catered, Limited Availability, Contract cost: £7,208.00 Contract length: 39 weeks, Size approx: 17.5m², Room with two single beds and a private bathroom with shower, sink and toilet. A wardrobe, bedside table, desk and chair, lamp and storage are provided per person. Duvet and pillows provided.Also see:What we provide in each hall[PDF]Table of bed sizes[PDF]Contract lengths and dates"
]

# Define patterns for extracting the required information
patterns = {
    'Room Type': r'H2: (.+)',
    'Contract Length': r'(\d+-week contract)',
    'Price Per Week': r'£[\d.]+( - £[\d.]+)?p/wk',
    'Catered': r'(Catered)',
    'Availability': r'(No Availability|Limited Availability)',
    'Size': r'Size approx: ([\d.]+m²( to [\d.]+m²)?)'
}

# Prepare a list to hold dictionaries for each room
rooms_data = []

# Extract room types from the first item in the list
room_types = re.search(patterns['Room Type'], rooms_content[0]).group(1).split(', ')

# Remove the room types from the first description to process it like the others
rooms_content[0] = re.sub(patterns['Room Type'], '', rooms_content[0])

# Iterate over each room's text, now that we have separated room types and details
for i, room_text in enumerate(rooms_content):
    room_info = {"Room Type": room_types[i] if i < len(room_types) else "Not Specified"}  # Handle case where there are more descriptions than types
    for key, pattern in patterns.items():
        if key != 'Room Type':  # Skip room type as it's already handled
            match = re.search(pattern, room_text)
            room_info[key] = match.group(0) if match else "Not Specified"
    rooms_data.append(room_info)

# Convert the list of dictionaries into a DataFrame
df_rooms = pd.DataFrame(rooms_data)

# Display the DataFrame to check
df_rooms
```

```python
import pandas as pd
import re


# Define a regex pattern to split the text into individual room sections
# This pattern might need adjustments based on actual data variations
pattern = r" (Shared|Private) bathroom"
room_sections = re.split(pattern, all_rooms_content)[1:]  # Split and ignore the header

# Reconstruct the sections properly after the split
room_sections = [f"{room_sections[i]} bathroom{room_sections[i + 1]}" for i in range(0, len(room_sections), 2)]

# List to hold room data
room_data = []

# Extract details for each room section
for room in room_sections:
    room_type_search = re.search(r"(Single room|Twin room|Twin en suite room)", room)
    contract_length = re.search(r"(\d+-week contract)", room)
    price_per_week = re.search(r"£[\d.,]+( - £[\d.,]+)?p/wk", room)
    catered = "Yes" if "Catered" in room else "No"
    availability = re.search(r"(No Availability|Limited Availability)", room)
    size = re.search(r"Size approx: ([\d.]+m²( to [\d.]+m²)?)", room)

    room_info = {
        "Room Type": room_type_search.group(0) if room_type_search else "Not specified",
        "Contract Length": contract_length.group(0) if contract_length else "Not specified",
        "Price Per Week": price_per_week.group(0) if price_per_week else "Not specified",
        "Catered": catered,
        "Availability": availability.group(0) if availability else "Not specified",
        "Size": size.group(0) if size else "Not specified"
    }
    room_data.append(room_info)

# Create DataFrame
df_rooms = pd.DataFrame(room_data)

# Display the DataFrame
df_rooms
```
