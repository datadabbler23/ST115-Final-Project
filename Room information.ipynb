```python
from bs4 import BeautifulSoup
import requests

all_hyperlinks = []

for page_num in range(1, 3):  #this way it goes thru both pages without having two diff links
    url = f"https://www.lse.ac.uk/student-life/accommodation/search-accommodation?collection=lse-accommodation&pageIndex={page_num}&sort=metaavailability"
    try:
        response = requests.get(url)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')
        accommodation_titles = soup.find_all('h2', class_='card__title')
        hyperlinks = []
        for title in accommodation_titles:
            hyperlink = title.find('a')['href']
            hyperlinks.append(hyperlink)
        all_hyperlinks.extend(hyperlinks)

    except Exception as e:
        print("An error occurred:", e)

print(all_hyperlinks)
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re

all_hyperlinks = ['http://www.lse.ac.uk/student-life/accommodation/halls/international-hall/home.aspx',]

all_data = []  

for url in all_hyperlinks:
    try:
        response = requests.get(url)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')
        
        
        accommodation_title_element = soup.find('h1', class_='heroBanner__title')
        accommodation_title = accommodation_title_element.get_text(strip=True) if accommodation_title_element else "Title Not Found"

        h2_tags = []  
        p_tags = []   


        rooms_divs = soup.find_all('div', class_='rooms')
        for div in rooms_divs:
            h2_tags.extend([h2.get_text(strip=True) for h2 in div.find_all('h2')])
            p_tags.extend([p.get_text(strip=True) for p in div.find_all('p')])

        shared_rooms = []
        private_room = []

        for item in p_tags:
            if item in ["Shared bathroom", "Private bathroom"] and private_room:
                shared_rooms.append(private_room)
                private_room = [item]
            else:
                private_room.append(item)

        if private_room:
            shared_rooms.append(private_room)

        for room in shared_rooms:
            if len(room) >= 8:
                contract_info = room[5]
                cost_match = re.search(r"Contract cost:\s*£([\d,\.]+)", contract_info)
                length_match = re.search(r"Contract length:\s*(\d+)\s*weeks", contract_info)

                contract_cost = cost_match.group(1) if cost_match else "Not available"
                contract_length = length_match.group(1) + " weeks" if length_match else "Not available"

                room_dict = {
                    "Accommodation Title": accommodation_title, 
                    "Room Types": h2_tags[0] if h2_tags else "Not specified",
                    "Bathroom Type": room[0],
                    "Contract Length": contract_length,
                    "Price per Week": room[2],
                    "Availability": room[4],
                    "Contract Cost": contract_cost,
                    "Size and Details": room[6]
                }
                all_data.append(room_dict)

    except Exception as e:
        print(f"An error occurred while scraping {url}: {e}")


df_all_rooms = pd.DataFrame(all_data)

df_all_rooms
```

```python
from bs4 import BeautifulSoup
import requests


url = "https://www.lse.ac.uk/student-life/accommodation/halls/rosebery-hall"


all_rooms_content = ""

try:
    response = requests.get(url)
    response.raise_for_status()  
    soup = BeautifulSoup(response.text, 'html.parser')
    

    rooms_divs = soup.find_all('div', class_='rooms')
    for div in rooms_divs:
        room_content = ""
        room_content += "H2: " + " ".join(h2.get_text(strip=True) for h2 in div.find_all('h2'))
        room_content += " P: " + " ".join(p.get_text(strip=True) for p in div.find_all('p'))
        all_rooms_content += room_content
    
except Exception as e:
    print("An error occurred:", e)

print(all_rooms_content)
```

```python
import pandas as pd
import re



pattern = r" (Shared|Private) bathroom"
room_sections = re.split(pattern, all_rooms_content)[1:]


room_sections = [f"{room_sections[i]} bathroom{room_sections[i + 1]}" for i in range(0, len(room_sections), 2)]


room_data = []


for room in room_sections:
    room_type_search = re.search(r"(Single room|Twin room|Twin en suite room)", room)
    contract_length = re.search(r"(\d+-week contract)", room)
    price_per_week = re.search(r"£[\d.,]+( - £[\d.,]+)?p/wk", room)
    catered = "Yes" if "Catered" in room else "No"
    availability = re.search(r"(No Availability|Limited Availability)", room)
    size = re.search(r"Size approx: ([\d.]+m²( to [\d.]+m²)?)", room)

    room_info = {
        "Room Type": room_type_search.group(0) if room_type_search else "Not specified",
        "Contract Length": contract_length.group(0) if contract_length else "Not specified",
        "Price Per Week": price_per_week.group(0) if price_per_week else "Not specified",
        "Catered": catered,
        "Availability": availability.group(0) if availability else "Not specified",
        "Size": size.group(0) if size else "Not specified"
    }
    room_data.append(room_info)


df_rooms = pd.DataFrame(room_data)
df_rooms
```
